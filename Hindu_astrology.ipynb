```python
import swisseph as swe
import pandas as pd
from datetime import datetime, timedelta

# Set the path to the Swiss Ephemeris files
swe.set_ephe_path('/path/to/ephemeris/files')  # Replace with the correct path

# Define date range
start_date = datetime(2023, 1, 1)  # Start date
end_date = datetime(2023, 10, 10)   # End date

# Define planets to include
planets = [
    swe.SUN, swe.MOON, swe.MARS, swe.MERCURY,
    swe.JUPITER, swe.VENUS, swe.SATURN,
    swe.MEAN_NODE  # Rahu (Mean Lunar Node)
]

planet_names = {
    swe.SUN: "Sun",
    swe.MOON: "Moon",
    swe.MARS: "Mars",
    swe.MERCURY: "Mercury",
    swe.JUPITER: "Jupiter",
    swe.VENUS: "Venus",
    swe.SATURN: "Saturn",
    swe.MEAN_NODE: "Rahu"
}

# Initialize an empty list to store data
data = []

# Generate planetary positions for each day in the date range
current_date = start_date
while current_date <= end_date:
    jd = swe.julday(current_date.year, current_date.month, current_date.day)
    row = {"Date": current_date.strftime("%Y-%m-%d")}
    
    # Get planetary positions
    for planet in planets:
        position, ret = swe.calc_ut(jd, planet)
        row[planet_names[planet]] = position[0]  # Store longitude in degrees
    
    data.append(row)
    current_date += timedelta(days=1)

# Convert data to a pandas DataFrame
df = pd.DataFrame(data)

# Save the dataset to a CSV file (optional)
df.to_csv("planetary_positions.csv", index=False)
```

```python
import swisseph as swe

# Set the path to Swiss Ephemeris data files
swe.set_ephe_path('/path/to/ephemeris/files')  # Replace with your ephemeris directory

# Input birth details
birth_date = (1995, 7, 10)  # YYYY, MM, DD
birth_time = (14, 30)  # HH, MM (24-hour format)
latitude = 28.6139  # Replace with your latitude
longitude = 77.2090  # Replace with your longitude
timezone_offset = 5.5  # IST offset from UTC in hours

# Compute Julian Day (required for Swiss Ephemeris calculations)
julian_day = swe.julday(birth_date[0], birth_date[1], birth_date[2], 
                        birth_time[0] + birth_time[1] / 60 - timezone_offset)

# Get planetary positions
planets = ['Sun', 'Moon', 'Mars', 'Mercury', 'Jupiter', 'Venus', 'Saturn', 'Rahu', 'Ketu']
planet_positions = {}

for i, planet in enumerate(planets, start=swe.SUN):
    position, _ = swe.calc_ut(julian_day, i)
    planet_positions[planet] = position[0]  # Longitude in degrees

# Print planetary positions
print("Planetary Positions (in degrees):")
for planet, position in planet_positions.items():
    print(f"{planet}: {position:.2f}°")

# Compute nakshatra (Lunar Mansion)
nakshatras = [
    "Ashwini", "Bharani", "Krittika", "Rohini", "Mrigashirsha", "Ardra", "Punarvasu",
    "Pushya", "Ashlesha", "Magha", "Purva Phalguni", "Uttara Phalguni", "Hasta", 
    "Chitra", "Swati", "Vishakha", "Anuradha", "Jyeshtha", "Mula", "Purva Ashadha", 
    "Uttara Ashadha", "Shravana", "Dhanishta", "Shatabhisha", "Purva Bhadrapada", 
    "Uttara Bhadrapada", "Revati"
]

moon_position = planet_positions['Moon']
nakshatra_index = int(moon_position // 13.3333)  # Each nakshatra spans 13°20'
current_nakshatra = nakshatras[nakshatra_index]

print(f"\nCurrent Nakshatra: {current_nakshatra}")

# Compute Vimshottari Dasha
dasha_lords = ["Ketu", "Venus", "Sun", "Moon", "Mars", "Rahu", "Jupiter", "Saturn", "Mercury"]
moon_degrees = moon_position % 13.3333
current_dasha = dasha_lords[int(moon_position // 120) % 9]
dasha_balance = 120 - (moon_degrees / 120 * 120)  # Years remaining in the current dasha

print(f"Current Dasha: {current_dasha} (Balance: {dasha_balance:.2f} years)")

# Compute planetary aspects
def compute_aspects(positions):
    aspects = {}
    for p1, pos1 in positions.items():
        for p2, pos2 in positions.items():
            if p1 != p2:
                aspect_angle = abs(pos1 - pos2) % 360
                aspects[f"{p1}-{p2}"] = aspect_angle
    return aspects

aspects = compute_aspects(planet_positions)
print("\nPlanetary Aspects (in degrees):")
for pair, angle in aspects.items():
    print(f"{pair}: {angle:.2f}°")

```

```python
import requests
from bs4 import BeautifulSoup
import time

base_url = "https://www.astro.com/astro-databank/Special:AllPages"

all_names = []

def get_names_from_page(url):

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    name_elements = soup.find_all('a', href=True, title=True)
    names = [name_element.text for name_element in name_elements if "/astro-databank/" in name_element['href']]
    return names

def scrape_all_pages():
    """Scrape names from all alphabetical pages."""
    global all_names
 
    response = requests.get(base_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    letter_links = soup.find('div', class_='mw-allpages-nav').find_all('a', href=True)
    letter_urls = ["https://www.astro.com" + link['href'] for link in letter_links]
    
    for letter_url in letter_urls:
        print(f"Scraping names from: {letter_url}")
        
        while letter_url:
            names = get_names_from_page(letter_url)
            all_names.extend(names)
            
            response = requests.get(letter_url)
            soup = BeautifulSoup(response.text, 'html.parser')
            next_page = soup.find('a', string="next 200")  
            letter_url = "https://www.astro.com" + next_page['href'] if next_page else None
            
            time.sleep(1)

    print("Scraping complete!")

scrape_all_pages()

with open("astro_names.txt", "w", encoding="utf-8") as file:
    for name in all_names:
        file.write(name + "\n")

print(f"Total names scraped: {len(all_names)}")
```
